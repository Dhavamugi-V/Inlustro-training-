```python
import seaborn as sns
plt.figure(figsize=(10,5))
sns.histplot(x["Age"],bins=30,kde=True)
plt.title("age distribution",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.scatterplot(data=x[["Age","Fare"]],palette="dark")
plt.title("scatter plot",fontsize=14)
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x[["Survived","Fare"]])
plt.title("boxplot")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Survived",data=x,palette="dark")
plt.title("survival count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x="Survived",y="Fare",data=x)
plt.title("violin plot for fare distribution",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(x["Age"],shade=True)
plt.title("density plot")
plt.show()
```
