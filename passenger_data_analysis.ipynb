```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/passenger data.csv")
data.head(5)
```

```python
data.info()
```

```python
data.shape
```

```python
data.drop(["Name","Cabin","SibSp"],axis=1)
```

```python
data.rename(columns={"Sex":"Gender"},inplace=True)
print(data)
```

```python
duplicate_rows_data=data[data.duplicated()]
print("number of duplicate rows",duplicate_rows_data.shape)
```

```python
data.dropna()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Fare',data=data,palette='coolwarm')
plt.title('price Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Gender',y='Survived',data=data,palette='dark')
plt.title("Survival Count by gender")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data["Age"],bins=30,kde=True)
plt.title("age distribution",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x="Survived",y="Age",hue="Gender",data=data)
plt.title("violin plot for:Age vs survival by gender",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Embarked', y='Fare', data=data, palette="dark")
plt.title("LINE PLOT",fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data["Age"],shade=True)
plt.title("density plot")
plt.show()
```
